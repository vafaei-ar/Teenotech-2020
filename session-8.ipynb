```
import numpy as np
import pandas as pd
```

```
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)#List')[3]
df = df.dropna()
```

```
df.sort_values('Population(1 July 2018)')
```

```
def correct1(x):
  try:
    return float(x[:-1].replace('−','-'))
  except:
    return 0
```

```
float('+0.43'),float('-0.43')
```

```
df['Change'] = df['Change'].apply(correct1)
```
