```
def return_primes2(x):
  result = []
  result.append(2)
  for i in range(3,x+1,2):
    is_prime = True
    for j in range(2,i):
      if i%j==0:
        is_prime = False
        break

    if is_prime==True:
      result.append(i)

  return result
```

```
primes = return_primes2(100000)
```

```
len(primes)
```

```
x = []
y = []
n = len(primes)
for i in range(n-1):
  x.append(i)
  y.append(primes[i+1]-primes[i])
```

```
plt.plot(x,y)
```
