```
def return_primes2(x):
  result = []
  print(2)
  for i in range(3,x+1,2):
    is_prime = True
    for j in range(2,i):
      if i%j==0:
        is_prime = False
        break

    if is_prime==True:
      result.append(i)

  return result
```

```
return_primes2(15)
```
